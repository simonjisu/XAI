```python
import torch

from lucent.optvis import render, param, transform, objectives
from lucent.modelzoo.util import get_model_layers
from lucent.modelzoo import resnet50, inceptionv1
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = resnet50(pretrained=True)
model.to(device).eval()
get_model_layers(model)[-5:]
```

```python
_ = render.render_vis(model, "labels:229", show_inline=True)
```

```python
obj = objectives.channel("fc", 593)
_ = render.render_vis(model, obj, show_inline=True)
```
